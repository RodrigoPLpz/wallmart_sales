# Importación de datos

```python
path = r'E:\\skillsForIndustry/project_wallmart/data/champagne_sales.csv'
sales_data = pd.read_csv(path, header=0, encoding='utf-8')
# Fuente consultada:
# https://raw.githubusercontent.com/SushmithaPulagam/TimeSeries_Auto-ARIMA/master/Champagne%20Sales.csv

path = r'E:\\skillsForIndustry/project_wallmart/data/Walmart_Store_sales.csv'
wall_data = pd.read_csv(path, header=0, encoding='utf-8')
```

```python
# sales_data.head()
# sales_data['Month'] = pd.to_datetime(sales_data['Month'], format='%d-%m-%Y').dt.strtime('%Y-%m-%d')
sales_data.set_index('Month',inplace=True)
sales_data = sales_data.rename_axis(index=None,columns=None)

# sales_data.index = sales_data.Month

sales_data.head()
```

```python
plt.rcParams["figure.figsize"] = (8,6)
sales_data.plot()
```

```python
from pmdarima.arima import ADFTest
adf_test = ADFTest(alpha = 0.05)
adf_test.should_diff(sales_data)
```

```python
train = sales_data[:85]
test = sales_data[-20:]
plt.rcParams["figure.figsize"] = (10,8)
plt.plot(train)
plt.plot(test)
```

```python
import pmdarima as pm
arima_model = pm.auto_arima(train, start_p=0,d=1,start_q=0,max_p=5,max_d=5,max_q=5,
                            start_P=0,D=1,start_Q=0,max_P=5,max_D=5,max_Q=5,m=12,
                            seasonal=True,error_Action='warn',trace=True,
                            supress_warnings = True, stepwise=True, random_state=20,
                            n_fits = 50)
```

```python
arima_model.summary()
```

```python
prediction = pd.DataFrame(arima_model.predict(n_periods=20),index=test.index)
prediction.columns = ['predicted_sales']
prediction
```

```python
plt.rcParams["figure.figsize"] = (10,8)
plt.plot(train,label='Training')
plt.plot(test,label='Test')
plt.plot(prediction,label='Predicted')
plt.legend(loc='upper left')
plt.show()
```

```python
from sklearn.metrics import r2_score
test['predicted_sales'] = prediction
r2_score(test['Champagne sales'], test['predicted_sales'])
```
