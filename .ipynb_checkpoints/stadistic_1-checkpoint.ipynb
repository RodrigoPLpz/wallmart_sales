# Importación de datos

```python
path = r'E:\\skillsForIndustry/project_wallmart/data/Walmart_Store_sales.csv'
wall_data = pd.read_csv(path, header=0, encoding='utf-8')
# Fuete consultada: https://www.kaggle.com/datasets/rutuspatel/walmart-dataset-retail
# How to Select a Model For Your Time Series Prediction Task [Guide]
# https://neptune.ai/blog/select-model-for-time-series-prediction-task
```

```python
wall_data.keys()
```

```python
wall_data_2 = wall_data.iloc[0:,[1,2]]
wall_data_2['Date'] = pd.to_datetime(wall_data['Date'],format="%d-%m-%Y", dayfirst=True).dt.strftime("%Y-%m-%d")
store_1 = wall_data_2.iloc[0:143,[0,1]]
store_1.set_index('Date',inplace=True)
store_1 = store_1.rename_axis(index=None, columns=None)
plt.rcParams["figure.figsize"] = (10,6)
store_1.plot();
```

```python
# store_1.isna().sum()
result = seasonal_decompose(store_1.Weekly_Sales,model='additive',period = 48)
# 48 es el número de semanas promedio
# store_1.shape <- (143,0)
# period debe ser la mitad de la cantidad de datos que tiene
# result = seasonal_decompose(store_1,model='multiplicative', period=12)
plt.rcParams["figure.figsize"] = (10,8)
result.plot(); 
```

```python
from statsmodels.graphics.tsaplots import plot_acf
plt.rcParams["figure.figsize"] = (8,5)
plot_acf(store_1.Weekly_Sales);
# eje x : peridos atras
```

```python
from statsmodels.graphics.tsaplots import plot_pacf
plt.rcParams["figure.figsize"] = (8,5)
plot_pacf(store_1.Weekly_Sales);
```

```python
from statsmodels.tsa.stattools import adfuller
adf, pval, usedlag, nobs, crit_vals, icbest =  adfuller(store_1.Weekly_Sales.values)
print('ADF test statistic:', adf)
print('ADF p-values:', pval)
print('ADF number of lags used:', usedlag)
print('ADF number of observations:', nobs)
print('ADF critical values:', crit_vals)
print('ADF best information criterion:', icbest)
```

```python
# Hay estacionalidad porque el p-values es menor de 0.05
```

#### AutoARIMA

```python
# !pip install pmdarima --user
import pmdarima as pm
from pmdarima.model_selection import train_test_split
```

```python
train, test = train_test_split(store_1.Weekly_Sales.values,train_size=120)
# train, test = train_test_split(store_1.values,train_size=100)
# Proporcionalidad 70 - 80
# todo
# --Segementacion el pasado mas reciente--
```

```python
model = pm.auto_arima(train, seasonal=True, m=52)
# m=52 weekly
# m=12 mounthy
# m=7 daily
# m = cuantos periodos se necesitan?
preds = model.predict(test.shape[0])
```

```python
# x = np.arange(store_1.shape[0])
# plt.plot(x[:140],train)
# plt.plot(x[140:],preds)
# plt.show()
x = np.arange(store_1.shape[0])
plt.plot(x[:120],train)
plt.plot(x[120:],preds)
plt.show()
```

#### Lineal Regression 

```python
# todo
# usar una segementacion en el train



# wall_data_2['Date'] = pd.to_datetime(wall_data['Date'],format="%d-%m-%Y", dayfirst=True).dt.strftime("%Y-%m-%d")
wall_data_3 = wall_data.iloc[0:143,[1,2]]
# wall_data_2.set_index('Date', inplace=True)
years =pd.DatetimeIndex(wall_data_3['Date']).year
months=pd.DatetimeIndex(wall_data_3['Date']).month
wall_data_3.set_index('Date', inplace=True)
wall_data_3 = wall_data_3.rename_axis(index=None, columns=None)
X=np.array([years,months]).T
```

```python
from sklearn.linear_model import LinearRegression

my_lr = LinearRegression()
my_lr.fit(X,wall_data_3.Weekly_Sales.values)

preds = my_lr.predict(X)
plt.rcParams["figure.figsize"] = (18,3)
plt.plot(wall_data_3.index, wall_data_3.Weekly_Sales.values)
plt.plot(wall_data_3.index,preds)
```

#### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

my_lr = RandomForestRegressor()
my_lr.fit(X,wall_data_3.Weekly_Sales.values)

preds = my_lr.predict(X)
# plt.rcParams["figure.figsize"] = (20,3)
plt.plot(wall_data_3.index, wall_data_3.Weekly_Sales.values)
plt.plot(wall_data_3.index,preds)
```
