```python
import numpy as np
import pandas as pd
```

```python
r2_ar_sem = [36.34,69.31,76.03,15.76,87.07,71.96,10.74,-47.37]
mape_ar_sem = [11.43,5.27,4.95,10.16,3.39,4.90,2.25,5.99]
r2__ar_men = [29.90,85.08,67.81,-3.68,96.07,87.41,-527.27,-323.33]
mape__ar_men = [9.01,3.09,4.45,1.53,1.86,2.93,3.59,6.83]
r2_pp_sem = [56.65,69.72,75.09,66.46,74.65,82.29,-600.33,-31.55]
mape_pp_sem = [9.6,4.62,4.10,4.47,5.14,5.48,9.13,5.58]
r2_pp_men = [21.73,11.18,24.21,2.79,6.36,51.21,-10345.67,-92.95]
mape_pp_men = [10.03,6.59,5.32,6.92,6.75,4.87,15.45,4.44]
tiendas = ['store_14','store_20','store_13','store_39','store_10','store_27','store_30','store_44']
```

```python
arima_semanal = {'store':tiendas,'r2':r2_ar_sem,'MAPE':mape_ar_sem}
prophet_semanal = {'store':tiendas,'r2':r2_pp_sem,'MAPE':mape_pp_sem}
arima_mes = {'store':tiendas,'r2':r2__ar_men,'MAPE':mape__ar_men}
prophet_mes = {'store':tiendas,'r2':r2_pp_men,'MAPE':mape_pp_men}
```

```python
arima_semanal = pd.DataFrame(arima_semanal)
prophet_semanal = pd.DataFrame(prophet_semanal)
arima_mes = pd.DataFrame(arima_mes)
prophet_mes = pd.DataFrame(prophet_mes)
```

```python
arima_semanal['tipo'] = list(np.repeat('ar_w', 8))
prophet_semanal['tipo'] = list(np.repeat('pp_w', 8))
arima_mes['tipo'] = list(np.repeat('ar_m', 8))
prophet_mes['tipo'] = list(np.repeat('pp_m', 8))
```

```python
prophet_semanal.sort_values(by = 'r2', ascending=False)
```

```python
arima_semanal.sort_values(by = 'r2', ascending=False)
```

```python
arima_mes.sort_values(by = 'r2', ascending=False)
```

```python
prophet_mes.sort_values(by = 'r2', ascending=False)
```

```python
resultados = pd.concat([arima_semanal,prophet_semanal,arima_mes,prophet_mes])
resultados = resultados.sort_values(by = 'r2', ascending=False)
```
